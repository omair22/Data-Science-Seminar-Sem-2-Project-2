```python
import time
import pickle
import json
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from newspaper import Article, Config
from tensorflow.keras.models import load_model
from webdriver_manager.chrome import ChromeDriverManager

# === Load Sentiment Model and Vectorizer ===
model = load_model("sentiment_model-lemma.h5")
with open("tfidf_vectorizer-lemma.pkl", "rb") as f:
    tfidf_vectorizer = pickle.load(f)
labels = ['Negative', 'Positive']

# === Define sources including Al Jazeera ===
sources = {
    "BBC": "bbc.com/news",
    "CNN": "cnn.com",
    "Fox": "foxbusiness.com",
    "AlJazeera": "aljazeera.com",
    "NDTV": "ndtv.com"
}


# === Set up Selenium WebDriver ===
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

config = Config()
config.browser_user_agent = "Mozilla/5.0"
config.request_timeout = 10

keyword = input("üîé Enter a topic to search: ")
articles_data = []

# === Loop through each source ===
for source_name, site_url in sources.items():
    query = f"{keyword} site:{site_url}"
    search_url = f"https://duckduckgo.com/?q={query.replace(' ', '+')}"
    driver.get(search_url)
    time.sleep(3)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    links = []
    for a in soup.find_all('a', href=True):
        href = a['href']
        if site_url in href and 'duckduckgo.com/l/' not in href and href.startswith('http'):
            if href not in links:
                links.append(href)
        if len(links) == 10:
            break

    for url in links:
        try:
            article = Article(url, config=config)
            article.download()
            article.parse()

            if source_name == "Fox":
                soup = BeautifulSoup(article.html, 'html.parser')
                script_tag = soup.find("script", {"type": "application/ld+json"})
                if script_tag:
                    json_ld = json.loads("".join(script_tag.contents))
                    title = json_ld.get("headline", "")
                    summary = json_ld.get("description", "")
                else:
                    title = article.title
            else:
                article.nlp()
                title = article.title

            combined_text = article.title + " " + article.text
            features = tfidf_vectorizer.transform([combined_text])
            prediction = model.predict(features.toarray())

            if prediction.shape[1] == 1:
                sentiment = labels[int(prediction[0][0] >= 0.5)]
            else:
                sentiment = labels[prediction.argmax()]

            articles_data.append({
                "Source": source_name,
                "Title": title,
                "Article": article.text,
                "Sentiment": sentiment
            })

        except:
            continue

driver.quit()

# === Create DataFrame ===
df = pd.DataFrame(articles_data)

# === Plot Sentiment Distribution per Source ===
if not df.empty:
    source_sentiment = df.groupby(['Source', 'Sentiment']).size().unstack().fillna(0)
    total_sentiment = df['Sentiment'].value_counts()

    source_sentiment.plot(kind='bar', figsize=(10, 6), title="Sentiment Distribution by Source")
    plt.xlabel("News Source")
    plt.ylabel("Number of Articles")
    plt.tight_layout()
    plt.show()

    total_sentiment.plot(kind='bar', figsize=(6, 4), title="Overall Sentiment Distribution")
    plt.xlabel("Sentiment")
    plt.ylabel("Total Articles")
    plt.tight_layout()
    plt.show()
else:
    print("‚ö†Ô∏è No articles successfully analyzed.")

```

```python
import pickle
import pandas as pd
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
import matplotlib.pyplot as plt

# === Load POS-tagging based model and tokenizer ===
model = load_model("sentiment_model-1000words-maxlen.h5")
with open("tokenizer-1000.pkl", "rb") as f:
    tokenizer = pickle.load(f)

# === Load DataFrame (manually provided for now)
# df = pd.read_csv("bbc_ndtv_combined.csv")  # Placeholder if loading from file

# Sample DataFrame structure (no actual data included)


# === Run POS-based sentiment prediction ===
MAXLEN = 1000  # Match with training

new_sentiments = []
for text in df["Article"]:
    seq = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(seq, maxlen=MAXLEN, padding='post')
    pred = model.predict(padded)
    sentiment = ['Negative', 'Positive'][pred.argmax()] if pred.shape[1] > 1 else ['Negative', 'Positive'][int(pred[0][0] >= 0.5)]
    new_sentiments.append(sentiment)

# Add new predictions to the DataFrame
df["POS_Sentiment"] = new_sentiments

# === Plot sentiment comparison ===
df.groupby(["Source", "POS_Sentiment"]).size().unstack().fillna(0).plot(kind="bar", figsize=(10, 6), title="POS-Tagged Sentiment by Source")
plt.ylabel("Article Count")
plt.tight_layout()
plt.show()

# Total POS-based sentiment
df["POS_Sentiment"].value_counts().plot(kind="bar", figsize=(6, 4), title="Overall POS-Tagged Sentiment")
plt.ylabel("Article Count")
plt.tight_layout()
plt.show()
```
